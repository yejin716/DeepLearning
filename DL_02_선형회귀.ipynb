```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8]) # 공부한 시간
y = np.array([81,93,91,97]) # 시험 점수
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# 기울기 w, 절편 b

#초기화

w = 0
b = 0

#학습률(learning rate)
#임의로 0.03

lr = 0.03

#몇 번 반복될지 설정
epochs = 2001
```

```python
#경사 하강법(sgd)

# x값이 총 몇개인지 센다
n = len(x)

# 경사 하강법 >> (편미분해서 오차가 가장 적은 기울기(w) 찾는법)
for i in range(epochs):
    y_pred = w * x + b #1차 선형 회귀
    error = y - y_pred #오차

    # w(가중치), b(bias 편향) 편미분
    w_diff = (2/n) * sum(-x * (error)) #오차(비용)함수를 w로 편미분한 값
    b_diff = (2/n) * sum(-(error)) # 오차(비용)함수를 b로 편미분한 값

    w = w - lr * w_diff #학습률을 곱해서 w 업데이트
    b = b - lr * b_diff #학습률을 곱해서 b 업데이트

    if i % 100 == 0:
        print("epochs = %.f, 기울기(w) = %.04f, 절편(b) = %.04f" %(i,w,b))
```

```python
# 결과 해석
# w = 2.3 / b = 79 >> y_pred(예상시험성적) = 2.3*(공부시간) + 79
```

```python
y_pred = w * x + b

#그래프 출력
plt.scatter(x,y)
plt.plot(x, y_pred, 'r')
plt.show()
```

다중 선형 회귀분석

```python
#독립변수 2개 (공부시간, 과외시간)

x1 = np.array([2,4,6,8]) # 공부시간
x2 = np.array([0,4,2,3]) # 과외시간
y = np.array([81,93,91,97]) # 점수
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(x1,x2,y)
plt.show()
```

```python
#변수 설정

#초기화
w1 = 0
w2 = 0
b = 0

#학습률 설정
lr = 0.01

#반복 횟수
epochs = 3001
```

```python
n = len(x1)

for i in range(epochs):
    y_pred = w1 * x1 + w2 * x2 + b
    error = y - y_pred

    w1_diff = (2/n) * sum(-x1 * (error))
    w2_diff = (2/n) * sum(-x2 * (error))
    b_diff = (2/n) * sum(-(error))

    w1 = w1 - lr * w1_diff
    w2 = w2 - lr * w2_diff
    b = b - lr * b_diff

    if i % 100 == 0:
        print("epochs = %.f, 기울기1(w1) = %.04f,  기울기2(w2) = %.04f, b = %.04f" %(i,w1,w2,b))
```

```python
#실제 점수와 예측 점수 출력

print('실제 점수: ', y)
print('예측 점수: ', y_pred)
```

# 딥러닝 기법 선형회귀

```python
#텐서플로우 활용

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2,4,6,8]) #공부시간
y = np.array([81,93,91,97]) #점수
```

```python
model = Sequential()
# 출력값 1, 입력 변수 1 , 분석방법 모델설정
model.add(Dense(1, input_dim=1, activation='linear'))
# 오차 수정(경사하강법(sgd)) / 오차정도확인(평균제곱오차(mse))
model.compile(optimizer='sgd', loss='mse')

#학습
model.fit(x,y,epochs=2000)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
#모델 테스트
hour = 7
prediction = model.predict([hour])

print('%.f 시간을 공부하시면, 예상점수는 %.02f 점 입니다.' %(hour, prediction))
```

# 딥러닝 기법 다중선형회귀

```python
x = np.array([[2,0],[4,4],[6,2],[8,3]]) # 학습시간/ 과외시간
y = np.array([81,93,91,97]) # 점수
```

```python
model = Sequential()
# 출력값 1, 입력 변수 1 , 분석방법 모델설정
model.add(Dense(1, input_dim=2, activation='linear'))
# 오차 수정(경사하강법(sgd)) / 오차정도확인(평균제곱오차(mse))
model.compile(optimizer='sgd', loss='mse')

#학습
model.fit(x,y,epochs=2000)
```

```python
#모델 테스트
hour = 7
tutoring_hour = 4
prediction = model.predict([[hour, tutoring_hour]])

print('%.f 시간을 공부하고 %.f시간의 과외를 받을 경우,  예상점수는 %.02f 점 입니다.' %(hour, tutoring_hour,prediction))
```
